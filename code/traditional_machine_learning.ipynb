```python
import matplotlib.pyplot as plt
import shap
import matplotlib.colors as mcolors
import pandas as pd
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score, average_precision_score, roc_curve, precision_recall_curve, auc

train = pd.read_excel('D:/研究生/Misdiagnosis/code/data/hybrid_machine_learning/train_without.xlsx')
train.dropna(how='any', inplace=True)
test = pd.read_excel('D:/研究生/Misdiagnosis/code/data/hybrid_machine_learning/test_without.xlsx')
test.dropna(how='any', inplace=True)

train_labels = train.iloc[:, -1]
test_labels = test.iloc[:, -1]
train.drop(train.columns[-1], axis=1, inplace=True)
test.drop(test.columns[-1], axis=1, inplace=True)

# # Define models
# models = [
#     ('Logistic Regression', LogisticRegression(max_iter=1000)),
#     ('SVM', SVC(kernel='linear', probability=True)),
#     ('Decision Tree', DecisionTreeClassifier()),
#     ('Random Forest', RandomForestClassifier()),
#     ('XGBoost', XGBClassifier())
# ]
#
#
# # Evaluate models
# for name, model in models:
#     model.fit(train, train_labels)
#     test_preds = model.predict(test)
#     test_prob_preds = model.predict_proba(test)[:, 1]
#
#     acc = accuracy_score(test_labels, test_preds)
#     auroc = roc_auc_score(test_labels, test_prob_preds)
#     auprc = average_precision_score(test_labels, test_prob_preds)
#
#     print(f'{name}: performance after oversampling:')
#     print(f'Accuracy = {acc:.2f}')
#     print(f'AUROC = {auroc:.2f}')
#     print(f'AUPRC = {auprc:.2f}')
#
#     # Create SHAP explainer and calculate SHAP values
#     explainer = shap.Explainer(model, train)
#     shap_values = explainer(test)
#
#     # Plot summary plot for top 10 features
#     shap.summary_plot(shap_values, test, max_display=5)
# model = SVC(kernel='linear', probability=True)
model = RandomForestClassifier()
model.fit(train, train_labels)
test_preds = model.predict(test)
test_prob_preds = model.predict_proba(test)[:, 1]

acc = accuracy_score(test_labels, test_preds)
auroc = roc_auc_score(test_labels, test_prob_preds)
auprc = average_precision_score(test_labels, test_prob_preds)

print(f'performance:')
print(f'Accuracy = {acc:.2f}')
print(f'AUROC = {auroc:.2f}')
print(f'AUPRC = {auprc:.2f}')

# calculate the ROC curve
fpr, tpr, thresholds = roc_curve(test_labels, test_prob_preds)

# calculate the PR curve
precision, recall, thresholds = precision_recall_curve(test_labels, test_prob_preds)

# Create a DataFrame for ROC curve
df_roc = pd.DataFrame({
    'FPR': fpr,
    'TPR': tpr,
})
# Write the ROC DataFrame to an Excel file
df_roc.to_excel("roc_data.xlsx", index=False)

# Create a DataFrame for PR curve
df_pr = pd.DataFrame({
    'Recall': recall,
    'Precision': precision,
})

# Write the PR DataFrame to an Excel file
df_pr.to_excel("pr_data.xlsx", index=False)



# # Set the font to Times New Roman
# plt.rcParams['font.family'] = 'Times New Roman'
#
# # Create SHAP explainer and calculate SHAP values
# explainer = shap.Explainer(model, train)
# shap_values = explainer(test)
#
# # Define a colormap
# cmap = mcolors.LinearSegmentedColormap.from_list("", ["#ffffd9", "#edf8b1", "#c7e9b4", "#7fcdbb", "#41b6c4", "#1d91c0", "#225ea8", "#253494", "#081d58"])
#
# # Summary plot
# shap.summary_plot(shap_values, test, plot_type='dot', color=cmap, max_display=10)
# # Get current figure and axes
# fig = plt.gcf()
# axes = plt.gca()
# # # Add vertical lines for SHAP values not equal to zero
# # for line in axes.get_lines():
# #     x = line.get_xdata()
# #     if x[0] != 0.0:  # if SHAP value not equal to zero
# #         plt.axvline(x=x[0], linestyle='dotted', color=line.get_color())
#
# # Save the plot
# plt.savefig("D:/研究生/Misdiagnosis/DSS/new_figure/fig4_a_shap_summary.png", dpi=1200)
# plt.show()
# plt.close()
# # Assuming your shap_values and X are already defined
# shap.plots.heatmap(shap_values)
# plt.savefig("D:/研究生/Misdiagnosis/DSS/new_figure/fig4_b_shap_heatmap.png", dpi=1200)
# plt.show()
```
